```python
%pip install gmpy2
%pip install cryptography
import gmpy2
import random
import timeit
import os
import binascii
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives.asymmetric import rsa
from cryptography.hazmat.primitives import serialization

def analyze_rsa_exponents(p, q):
    """
    Calcule et affiche les tailles des exposants privés RSA 
    pour des exposants publics donnés et aléatoires.
    """
    # 1. Conversion en type mpz pour la rapidité de gmpy2
    p_mpz = gmpy2.mpz(p)
    q_mpz = gmpy2.mpz(q)
    
    # 2. Calculs de base
    n = p_mpz * q_mpz
    phi_n = (p_mpz - 1) * (q_mpz - 1)
    
    # gmpy2.mpz possède la méthode bit_length() de manière native
    n_bits = n.bit_length() 
    
    # 3. Préparation des exposants à tester
    exposants_a_tester = [3, 17, 65537] # 65537 = 2^16 + 1
    
    # Génération de 3 exposants aléatoires valides (premiers avec phi_n)
    compteur_aleatoire = 0
    while compteur_aleatoire < 3:
        # e_rand choisi aléatoirement entre 65538 et phi_n - 1
        e_rand = gmpy2.mpz(random.randint(65538, int(phi_n) - 1))
        # Vérification mathématique stricte pour RSA : gcd(e, phi_n) == 1
        if gmpy2.gcd(e_rand, phi_n) == 1:
            exposants_a_tester.append(int(e_rand))
            compteur_aleatoire += 1

    # 4. Affichage de l'en-tête
    print(f"taille de n (bits) : {n_bits}\n")
    print(f"{'exposant public (e)':<22} | {'taille de e (bits)':<18} | {'exposant privé (d)':<22} | {'taille de d (bits)'}")
    print("-" * 26 + "+" + "-" * 37 + "+" + "-" * 20)
    
    # 5. Calcul et remplissage du tableau
    for e in exposants_a_tester:
        e_mpz = gmpy2.mpz(e)
        
        # Gérer le cas où les exposants fixes choisis (3 ou 17) ne sont pas
        # premiers avec phi(n) par malchance avec p et q
        if gmpy2.gcd(e_mpz, phi_n) != 1:
            print(f"{e:<25} | {'Erreur : PGCD(e, phi(n)) != 1':<35} | -")
            continue
            
        # Calcul ultra-rapide de l'inverse modulaire
        d = gmpy2.invert(e_mpz, phi_n)
        d_bits = d.bit_length()
        e_bits = e_mpz.bit_length()
        
        # Troncature visuelle de d pour ne pas casser l'affichage du tableau 
        # si le nombre est gigantesque
        d_str = str(d)
        if len(d_str) > 30:
            d_str = d_str[:8] + "..." + d_str[-8:]
        e_str = str(e_mpz)
        if len(e_str) > 30:
            e_str = e_str[:8] + "..." + e_str[-8:]
            
        print(f"{e_str:<22} | {e_bits:<18} | {d_str:<22} | {d_bits}")



def get_rsa_triplet(key_size):
    """Génère m, e, d, n pour une taille donnée."""
    print(f"\n--- Génération clé {key_size} bits... ---")
    private_key = rsa.generate_private_key(
        public_exponent=65537,
        key_size=key_size,
        backend=default_backend()
    )
    priv_numbers = private_key.private_numbers()
    pub_numbers = priv_numbers.public_numbers
    
    n = pub_numbers.n
    e = pub_numbers.e
    d = priv_numbers.d
    
    # Message aléatoire m < n
    m = random.getrandbits(key_size - 1)
    
    return m, e, d, n

def benchmark_separate():
    key_sizes = [1024, 2048, 3072, 4096]
    
    # Paramètres d'itération on réalise 100 calculs pour avoir un temps moyen fiable:
    # Chiffrement très rapide -> on pourrait se permettre plus d'itérations
    iter_enc = 100 
    # Déchiffrement lent -> attention à ne pas proposer un trop grand nombre pour ne pas attendre 10 min
    iter_dec = 100     

    # En-tête du tableau
    header = f"| {'Opération':<15} | {'Taille':<6} | {'Python (sec/100)':<15} | {'gmpy2 (sec/100)':<12} | {'Gain':<8} |"
    print("-" * len(header))
    print(header)
    print("-" * len(header))

    for size in key_sizes:
        m, e, d, n = get_rsa_triplet(size)
        
        # --- Conversion gmpy2 (Hors chrono) ---
        m_g = gmpy2.mpz(m)
        e_g = gmpy2.mpz(e)
        d_g = gmpy2.mpz(d)
        n_g = gmpy2.mpz(n)
        
        # Pré-calcul du chiffré pour tester le déchiffrement ensuite
        c = pow(m, e, n)
        c_g = gmpy2.powmod(m_g, e_g, n_g)

        # ==========================================
        # 1. TEST CHIFFREMENT (Exposant e = 65537)
        # ==========================================
        t_enc_py = timeit.timeit(lambda: pow(m, e, n), number=iter_enc)
        t_enc_gm = timeit.timeit(lambda: gmpy2.powmod(m_g, e_g, n_g), number=iter_enc)
        ratio_enc = t_enc_py / t_enc_gm

        print(f"| {'CHIFFREMENT':<15} | {size:<6} | {t_enc_py:.5f}{' '*8} | {t_enc_gm:.5f}{' '*5} | x{ratio_enc:.2f}    |")

        # ==========================================
        # 2. TEST DÉCHIFFREMENT (Exposant d ~ n)
        # ==========================================
        t_dec_py = timeit.timeit(lambda: pow(c, d, n), number=iter_dec)
        t_dec_gm = timeit.timeit(lambda: gmpy2.powmod(c_g, d_g, n_g), number=iter_dec)
        ratio_dec = t_dec_py / t_dec_gm
        
        print(f"| {'DÉCHIFFREMENT':<15} | {size:<6} | {t_dec_py:.5f}{' '*8} | {t_dec_gm:.5f}{' '*5} | x{ratio_dec:.2f}    |")
        print("-" * len(header))


def int_to_bytes(i):
    # i might be a gmpy2 big integer; convert back to a python int
    i = int(i)
    # Assurer une taille minimum d'octets
    length = (i.bit_length() + 7) // 8
    return i.to_bytes(length, byteorder='big')

def bytes_to_int(b):
    return int.from_bytes(b, byteorder='big')

def main():
    public_key_file = None
    private_key_file = None
    public_key = None
    private_key = None

    while True: 
        print("\nSimple RSA Crypto")
        print("-----------------")
        print(f"\tprivate key file: {private_key_file}")
        print(f"\tpublic key file: {public_key_file}")
        print("\t1. Encrypt message")
        print("\t2. Decrypt message")
        print("\t3. Load public key file")
        print("\t4. Load private key file")
        print("\t5. Create and load new public and private key files.")
        print("\t6. Quit. \n")
        
        choice = input(">> ")

        if choice == '1':
            if not public_key:
                print("\nNo public key loaded\n")
            else:
                message = input("\nPlaintext: ").encode()
                message_as_int = bytes_to_int(message)
                cipher_as_int = simple_rsa_encrypt(message_as_int, public_key)
                cipher = int_to_bytes(cipher_as_int)
                print("\nCiphertext (hexlified): {}\n".format(binascii.hexlify(cipher).decode()),"\n")
                #

        elif choice == '2':
            if not private_key:
                print("\nNo private key loaded\n")
            else:
                cipher_hex = input("\nCiphertext (hexlified): ").encode()
                try:
                    cipher = binascii.unhexlify(cipher_hex)
                    cipher_as_int = bytes_to_int(cipher)
                    decrypted_int = simple_rsa_decrypt(cipher_as_int, private_key)
                    message = int_to_bytes(decrypted_int)
                    print("\nPlaintext: {}\n".format(message.decode(errors='ignore')))
                    print("\nPlaintext (hex) : {}\n".format(binascii.hexlify(cipher)))
                except Exception as e:
                    print(f"\nError decrypting: {e}\n")

        elif choice == '3':
            public_key_file_temp = input("\nEnter public key file: ")
            if not os.path.exists(public_key_file_temp):
                print(f"File {public_key_file_temp} does not exist.")
            else:
                with open(public_key_file_temp, "rb") as public_key_file_object:
                    public_key = serialization.load_pem_public_key(
                        public_key_file_object.read(),
                        backend=default_backend()
                    )
                    public_key_file = public_key_file_temp
                    print("\nPublic Key file loaded.\n")
                    
                    # unload private key if any just to be clean
                    # (optional depending on desired behavior, kept logic from original)
                    private_key_file = None
                    private_key = None

        elif choice == '4':
            private_key_file_temp = input("\nEnter private key file: ")
            if not os.path.exists(private_key_file_temp):
                print(f"File {private_key_file_temp} does not exist.")
            else:
                with open(private_key_file_temp, "rb") as private_key_file_object:
                    private_key = serialization.load_pem_private_key(
                        private_key_file_object.read(),
                        backend=default_backend(),
                        password=None
                    )
                    private_key_file = private_key_file_temp
                    print("\nPrivate Key File loaded.\n")

                    # Deduce public key from private key
                    public_key = private_key.public_key()
                    public_key_file = None  # Reset filename as it's derived, not loaded from file

        elif choice == '5':
            private_key_file_temp = input("\nEnter a file name for new private key: ")
            public_key_file_temp = input("\nEnter a file name for new public key: ")
            
            if os.path.exists(private_key_file_temp) or os.path.exists(public_key_file_temp):
                print("One of the files already exists. Aborting to prevent overwrite.")
            else:
                # Generation
                print("Generating key pair...")
                private_key = rsa.generate_private_key(
                    public_exponent=65537,
                    key_size=2048,
                    backend=default_backend()
                )
                public_key = private_key.public_key()

                # Serialization
                private_key_bytes = private_key.private_bytes(
                    encoding=serialization.Encoding.PEM,
                    format=serialization.PrivateFormat.TraditionalOpenSSL,
                    encryption_algorithm=serialization.NoEncryption()
                )
                
                public_key_bytes = public_key.public_bytes(
                    encoding=serialization.Encoding.PEM,
                    format=serialization.PublicFormat.SubjectPublicKeyInfo
                )

                # Writing to files
                with open(private_key_file_temp, "wb+") as private_key_file_obj:
                    private_key_file_obj.write(private_key_bytes)
                
                with open(public_key_file_temp, "wb+") as public_key_file_obj:
                    public_key_file_obj.write(public_key_bytes)

                # Update state
                public_key_file = public_key_file_temp
                private_key_file = private_key_file_temp
                print("\nNew keys generated and loaded.\n")

        elif choice == '6':
            print("\n\nTerminating.\n")
            break
        
        else:
            print(f"\n\nUnknown option {choice}.\n")
```
