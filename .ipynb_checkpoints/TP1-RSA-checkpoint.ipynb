### 💡 Rappel : Les formats de données en Python

En cryptographie, la donnée change constamment de forme. Avant de chiffrer, il est crucial de comprendre ce que l'on manipule :

* **Unicode (Texte / `str`)** : Ce sont les caractères que nous lisons (ex: `"Bonjour"`). C'est une abstraction pour les humains.
* **Bytes / Octets (`bytes`)** : C'est le langage brut de la machine, les bits sont traités par série de 8. Certains octets correspondent à des caractères unicode, d'autres ne s'affichent pas. **Les algorithmes de cryptographie ne calculent que sur des octets.** Pour indiquer que des caractères sont des octets, python utilise le préfixe `b"Hello World"`. 
* **Hexadécimal** : C'est une représentation mathématique en base 16 (chiffres de 0 à 9, et lettres de a à f). C'est la bonne façon d'afficher des octets à l'écran, pour éviter aux développeurs de lire des kilomètres de 0 et de 1.

### 🧮L'Arithmétique Modulaire pour RSA : Python Natif vs `gmpy2`

L'algorithme RSA repose entièrement sur l'arithmétique modulaire (les calculs modulo $n$) avec des nombres entiers gigantesques (souvent 1024 ou 2048 bits). Bien que Python gère nativement les entiers de taille arbitraire, il est parfois nécessaire, pour des raisons de performance, de s'appuyer sur des bibliothèques optimisées en C comme `gmpy2` (basée sur la bibliothèque GNU Multiple Precision).

Voici comment réaliser les opérations fondamentales de l'arithmétique modulaire selon ces deux approches.

### 1. Avec les fonctions natives de Python

Python intègre les outils mathématiques de base, suffisants pour comprendre et implémenter RSA sur des clés de petite taille à visée pédagogique.

* **Somme et Produit modulo $n$ :**
    On utilise simplement l'opérateur modulo `%`.
    ```python
    somme_mod = (a + b) % n
    produit_mod = (a * b) % n
    ```
</br>
* **Division euclidienne (Quotient et Reste) :**
    La fonction native `divmod()` est optimisée pour renvoyer les deux valeurs simultanément.
    ```python
    q, r = divmod(a, b) # a = b * q + r
    ```
</br>
* **Inverse modulaire (trouver $x$ tel que $a \cdot x \equiv 1 \pmod n$) :**
    Depuis Python 3.8, la fonction native `pow()` permet de calculer l'inverse modulaire très efficacement en passant `-1` comme exposant.
    ```python
    inverse = pow(a, -1, n)
    ```
</br>
* **Coefficients de Bézout (Algorithme d'Euclide Étendu) :**
    Il n'existe pas de fonction native en Python pour extraire les coefficients de Bézout ($u$ et $v$ tels que $a \cdot u + b \cdot v = \text{pgcd}(a,b)$).

---

### 2. Avec le module `gmpy2` (Haute Performance)

Pour des applications cryptographiques réelles ou pour accélérer drastiquement vos temps de calcul dans ce TP, le module `gmpy2` est l'outil de référence. Il utilise le type `mpz` (multiple precision zero) pour ses entiers.

*Note : Assurez-vous d'avoir installé le module (`pip install gmpy2`) et de l'importer.*
```python
import gmpy2
# Conversion des entiers standards en type optimisé mpz
a, b, n = gmpy2.mpz(a), gmpy2.mpz(b), gmpy2.mpz(n)
```

* **Somme et produits modulo $n$ :**
Vous pouvez utiliser les opérateurs standards (ils sont surchargés et rapides si les variables sont des mpz), ou les fonctions dédiées.

    ```python
somme_mod = gmpy2.f_mod(a+b,n)
produit_mod = gmpy2.f_mod(a*b,n)
    ```

* **Division euclidienne (quotient et reste) :** la fonction `t_divmod` renvoie un tuple contenant le quotient et le reste.
    ```python
q,r = gmpy2.t_divmod(a,b)
    ```

* **Inverse modulaire :** très performante, la fonction `invert`gère ce calcul fondamental pour la création de la clé privée RSA.
  ```python
  inverse = gmpy2.invert(a,n)
  ```
  </br>
* **Coefficients de Bézout :** contrairement à python natif, `gmpy2` propose une fonction intégrée extrêmement rapide pour calculer à la fois le PGCD et les coefficients de Bézout.
  ```python
    pgcd,u,v = gmpy2.gcdext(a,b)
    ```

## RSA

L'algorithme RSA repose sur la difficulté de la factorisation d'un grand nombre entier $n$. La clé publique est $(e,n)$ et la clé privée $(d,p,q)$. Rappelons que :
$$ n = p\times q \qquad \text{ et } \qquad \varphi(n) = (p-1)\times (q-1)$$

### Équations Fondamentales
* **Chiffrement :** $$y \equiv x^e \pmod n$$
* **Déchiffrement :** $$x \equiv y^d \pmod n$$
* **Relation entre $e$ et $d$ :**  $$  \overline{d} \quad \text{ est l'inverse de $\quad \overline{e}\quad$ dans } \quad \mathbb{Z} / \varphi(n) \mathbb{Z} $$
</div>

</br>

**✖️ La propriété multiplicative de RSA (Homomorphisme)**

Le cryptosystème RSA possède une particularité algébrique très intéressante : il est **partiellement homomorphe par rapport à la multiplication**. 

Concrètement, cela signifie que le produit de deux messages chiffrés est égal au chiffrement du produit des messages clairs originaux. 

Mathématiquement, si l'on note $E(m)$ la fonction de chiffrement dans $\mathbb{Z} / n \mathbb{Z}$ avec la clé publique $(e, n)$ :
$$E(\overline{m_1}) \cdot E(\overline{m_2})= (\overline{m_1}^e ) \cdot (\overline{m_2}^e ) = (\overline{m_1 \cdot m_2})^e \equiv E(\overline{m_1} \cdot \overline{m_2})$$

</div>

### Pourquoi choisir un $e$ petit ?
Dans les standards réels (comme ceux du NIST), on utilise souvent un exposant public parmi les trois valeurs suivantes : $3, 17, 65537 = 2^{16} + 1$. Ce choix n'est pas aléatoire :
1. **Optimisation :** Un $e$ petit avec peu de bits à `1` (poids de Hamming faible) permet de réduire drastiquement le nombre de multiplications lors de l'exponentiation modulaire.
2. **Conséquence :** En contrepartie, la clé privée $d$ résultante est un grand entier, ce qui protège contre les attaques brute-force mais qui a aussi une conséquence en terme d'efficacité. En raison d'un exposant privé très long, le déchiffrement et la signature sont beaucoup plus longs que le chiffrement et l'authentification (qui utilisent la clé publique). On peut montrer que pour garantir la sécurité de l'algorithme, la clé privée doit avoir une taille d'au moins $0.3t$ bits où $t$ est la longueur en bits du module $n$. 

### 🔍 Analyse de la taille des clés RSA (`analyze_rsa_exponents`)

Cette fonction va nous donner un aperçu des tailles de clé RSA. 

**Ce que fait la fonction :**
* **Initialisation :** Elle prend en entrée deux nombres premiers $p$ et $q$, puis calcule le module $n$ et affiche sa taille totale en bits.
* **Génération des exposants :** Elle simule la création de clés en utilisant des exposants publics $e$ standards et très petits (3, 17 et $2^{16}+1$), ainsi que quelques exposants générés aléatoirement.
* **Comparaison visuelle :** Elle affiche un tableau récapitulatif mettant en évidence les tailles respectives de $e$ et de son inverse modulaire $d$.

**L'objectif pédagogique :**
Ce tableau vous permet d'observer que même si l'on choisit un exposant public $e$ minuscule (pour accélérer volontairement le chiffrement), l'exposant privé $d$ calculé conservera toujours un taille de l'ordre de grandeur de $n$. C'est cette asymétrie de taille qui garantit la robustesse de la clé privée face aux attaques par force brute !

### Librairie cryptography et module hazmat. 

La librairie `cryptography` de Python que nous avons commencé à utiliser contient toutes les fonctions utiles en cryptographie implémentée de manière sécurisée : c'est la librairie standard pour les utilisateurs de Python. Cependant nous utilisons le module `hazmat` pour Hazardous Material parce que nous ne souhaitons pas utiliser ces fonctions dans des contextes réels mais observer leur fonctionnement en détail, y compris avec de mauvais paramétrages. Ce module, qui porte bien son nom ne devrait **jamais** être utilisé dans de vraies applications. Pour citer la documentation, 
>"You should ONLY use it if you’re 100% absolutely sure that you know what you’re doing because this module is full of land mines, dragons, and dinosaurs with laser guns.".

Mais même avec hazmat, cryptography ne vous laisse pas faire n'importe quoi. Vous ne pouvez par exemple pas générer de clé RSA pour un module de 512 bits via la fonction ```rsa.generate_private_key()``` (que se passe-t-il si vous essayez?). Vous ne pouvez pas non plus l'utiliser pour chiffrer et déchiffrer avec un RSA déterminsite : la librairie ne vous autorisera pas à utiliser RSA sans padding. 

````python
#Generate a private key

private_key = rsa.generate_private_key(
    public_exponent = 65537, 
    key_size = 1024, 
    backend=default_backend()
)

#Extract the public key from the private key
public_key = private_key.public_key()

# Extraction des nombres privés
private_numbers = private_key.private_numbers()

# Affichage des valeurs
print(f"p (premier facteur) = {private_numbers.p}")
print(f"q (second facteur)  = {private_numbers.q}")
print(f"d (exposant privé)  = {private_numbers.d}")

# Pour info, on peut vérifier que n = p * q
n = private_numbers.p * private_numbers.q
print(f"n (module) calculé  = {n}")

# Comparaison avec le module stocké dans la clé publique
print(f"n (module) officiel = {private_numbers.public_numbers.n}")
````

````python
def simple_rsa_encrypt(m, publickey):
    numbers = publickey.public_numbers()
    return gmpy2.powmod(m, numbers.e, numbers.n)

def simple_rsa_decrypt(c, privatekey):
    numbers = privatekey.private_numbers()
    return gmpy2.powmod(c, numbers.d, numbers.public_numbers.n)



if __name__ == '__main__':
    main()
````
